```python
import requests
import pandas as pd
import time
import json
import datetime
import plotly.express as px
from dataclasses import dataclass
from datetime import datetime
import numpy as np
from scipy import stats
from statsmodels.distributions.empirical_distribution import ECDF
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tvls = requests.get('https://api.llama.fi/protocol/Uniswap-v3').json()['chainTvls']
df = None
for key in tvls.keys():
    tmp = pd.DataFrame(tvls[key]['tvl'])
    
    if df is None:
        df = tmp
        continue

    df = df.merge(tmp, on='date', suffixes=('', key), how='outer')
    
df['date'] = df.date.apply(lambda x: datetime.utcfromtimestamp(x))
cols = list(df.columns)
cols.remove('date')
```

```python
df.describe()
```

```python
px.bar(df[df.date > '2022-01-01'], x='date', y=cols)
```

Как видно из графика выше, за последний год tvl uniswap-v3 в основном приходится на сеть eth.

```python
request_url = "https://api.coingecko.com/api/v3/coins/uniswap/market_chart?vs_currency=usd&days=max&interval=daily"
data = requests.get(request_url).json()
```

```python
items = []
for (p, mc, tv) in zip(data['prices'], data['market_caps'], data['total_volumes']):
    item = [p[0], p[1], mc[1], tv[1]]
    items.append(item)

df = pd.DataFrame(items)
```

```python
df.columns = ['ts', 'prices', 'market_caps', 'total_volumes']
df['ts'] = df.ts.apply(lambda t: datetime.utcfromtimestamp(t / 1000))
```

```python
df.describe()
```

```python
fig = px.line(x=df['ts'], y=df['prices'])
fig.show()
```

Как видно из графика выше в поселедний год наблюдается тенденция падения цен на UNI coins. Пик цен приходится на 5-е мая 2021 года и составляет 42.88 \$ за 1 Uniswap coin.  

```python
stats.kstest(df['prices'], stats.norm.cdf)
```

Данные не распределены нормально.

Как видно из теста Колмогорова-Смирнова - распределение цен не является нормальным. 

Нарисуем графики pdf и cdf.

```python
ecdf = ECDF(df['prices'])
plt.plot(ecdf.x,ecdf.y)
```

```python
sns.kdeplot(data=df, x="prices")
```

```python
fig = px.line(x=df['ts'], y=df['total_volumes'])
fig.show()
```
